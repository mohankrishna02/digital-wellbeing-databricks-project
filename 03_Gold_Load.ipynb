```python
from pyspark.sql.functions import *

silver_df = spark.table(
    "workspace.default.silver_digital_wellbeing"
)

display(silver_df)
```

```python
#Yearly trends (How are screen time, wellbeing, anxiety and depression changing over time?)

yearly_trends = (
    silver_df
    .groupBy("year")
    .agg(
        round(avg("avg_screen_time_hours"),2).alias("avg_screen_time"),
        round(avg("avg_social_media_hours"),2).alias("avg_social_media"),
        round(avg("avg_sleep_hours"),2).alias("avg_sleep"),
        round(avg("avg_anxiety_score"),2).alias("avg_anxiety"),
        round(avg("avg_depression_score"),2).alias("avg_depression"),
        round(avg("avg_digital_wellbeing"),2).alias("avg_wellbeing"),
        round(avg("avg_mental_health_risk"),2).alias("avg_risk")
    )
)

display(yearly_trends)

yearly_trends.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable(
        "workspace.default.gold_yearly_trends"
    )
```

```python
#Country Summary (Which countries have the highest risk and lowest wellbeing?)

country_summary = (
    silver_df
    .groupBy(
        "country",
        "continent",
        "income_group"
    )
    .agg(
        round(avg("avg_screen_time_hours"),2).alias("avg_screen_time"),
        round(avg("avg_sleep_hours"),2).alias("avg_sleep"),
        round(avg("avg_anxiety_score"),2).alias("avg_anxiety"),
        round(avg("avg_depression_score"),2).alias("avg_depression"),
        round(avg("avg_digital_wellbeing"),2).alias("avg_wellbeing"),
        round(avg("avg_mental_health_risk"),2).alias("avg_risk")
    )
)

display(country_summary)

country_summary.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable(
        "workspace.default.gold_country_summary"
    )

print("Table loaded successfully")
```

```python
#Income Group Analysis (How do income groups compare?)

income_group_analysis = (
    silver_df
    .groupBy("income_group")
    .agg(
        round(avg("avg_screen_time_hours"),2).alias("avg_screen_time"),
        round(avg("avg_sleep_hours"),2).alias("avg_sleep"),
        round(avg("avg_anxiety_score"),2).alias("avg_anxiety"),
        round(avg("avg_depression_score"),2).alias("avg_depression"),
        round(avg("avg_digital_wellbeing"),2).alias("avg_wellbeing"),
        round(avg("avg_mental_health_risk"),2).alias("avg_risk")
    )
)

display(income_group_analysis)

income_group_analysis.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable(
        "workspace.default.gold_income_group_analysis"
    )

print("Table loaded successfully")
```

```python
#Continent Analysis

continent_summary = (
    silver_df
    .groupBy("continent")
    .agg(
        round(avg("avg_screen_time_hours"),2).alias("avg_screen_time"),
        round(avg("avg_sleep_hours"),2).alias("avg_sleep"),
        round(avg("avg_digital_wellbeing"),2).alias("avg_wellbeing"),
        round(avg("avg_mental_health_risk"),2).alias("avg_risk")
    )
)

display(continent_summary)

continent_summary.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable(
        "workspace.default.gold_continent_summary"
    )

print("Table Loaded Sucessfully")
```
